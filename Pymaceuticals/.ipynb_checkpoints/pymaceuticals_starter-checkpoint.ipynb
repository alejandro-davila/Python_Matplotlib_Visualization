```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data=pd.merge(mouse_metadata, study_results, how="right",on="Mouse ID")


# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
mice_count=len(combined_data["Mouse ID"].unique())
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_data.loc[combined_data.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_df=combined_data.loc[combined_data['Mouse ID'] == "g989",:]
duplicate_mice_df.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df=combined_data.loc[combined_data['Mouse ID'].isin(duplicate_mice)==False]
clean_df                      
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice = clean_df['Mouse ID'].value_counts()
number_of_mice = len(clean_mice)
number_of_mice
```

## Summary Statistics

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
    # mean, median, variance, standard deviation, and SEM of the tumor volume. 
    # Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
    
regimen_tumor_mean = clean_df.groupby('Drug Regimen').mean(numeric_only=True)["Tumor Volume (mm3)"]
regimen_tumor_median = clean_df.groupby('Drug Regimen').median(numeric_only=True)["Tumor Volume (mm3)"]
regimen_tumor_var = clean_df.groupby ('Drug Regimen').var(numeric_only=True)["Tumor Volume (mm3)"]
regimen_tumor_std = clean_df.groupby ('Drug Regimen').std(numeric_only=True)["Tumor Volume (mm3)"]
regimen_tumor_sem = clean_df.groupby ('Drug Regimen').sem(numeric_only=True)["Tumor Volume (mm3)"]


# Assemble the resulting series into a single summary DataFrame.
summary_stat_summary = pd.DataFrame({'Mean of Tumor-Volume':regimen_tumor_mean,
                                   'Median of Tumor-Volume':regimen_tumor_median,
                                 'Variance of Tumor-Volume':regimen_tumor_var,
                       'Standard deviation of TumorVolume':regimen_tumor_std,
                                      'Std Err of Tumor-Volume':regimen_tumor_sem})
summary_stat_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_summary = clean_df.groupby(['Drug Regimen'])[["Tumor Volume (mm3)"]].agg(['mean','median','var','std','sem'])
agg_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
timepoint_max = combined_data.groupby(["Mouse ID"]).max()
mouse_index_reset = timepoint_max.reset_index()


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = mouse_index_reset[['Mouse ID','Timepoint']].merge(combined_data,on=['Mouse ID','Timepoint'],how="left")
merged_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset
    # Determine outliers using upper and lower bounds
def get_outliers(abc):
    regimen_stats = merged_data.loc[merged_data["Drug Regimen"] == abc]['Tumor Volume (mm3)']
                                                           
    quartiles = regimen_stats.quantile([.25,.5,.75])
    upperq = quartiles[0.25]
    lowerq = quartiles[0.75]
    iqr = lowerq - upperq
    lower_bound = upperq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    subset=merged_data.iloc[range(1,249)]
    outliers = regimen_stats.loc[(regimen_stats<lower_bound) | (regimen_stats>upper_bound)]
    print(f"{abc}'s potential outliers:{outliers}")                           
    return regimen_stats                         

cap = get_outliers("Capomulin")
ram = get_outliers("Ramicane")
inf = get_outliers("Infubinol")
cef = get_outliers("Ceftamin")
```
